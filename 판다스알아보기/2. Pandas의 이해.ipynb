```python
# 새 컬럼 추가하여 데이터 만들기
# loc을 사용하여 논리검색을 행 칸에 넣고 열 칸에는 새로운 컬럼 이름을 넣음. 데이터는 오른쪽에 엏음.
tips.loc[tips['tip']>3.6,'level'] = 'good'
tips.loc[(tips['tip']<=3.6)&(tips['tip']>=2),'level'] = 'soso'
tips.loc[tips['tip']<2,'level'] = 'bad'
print(tips['level'].isnull().sum())
tips
```

```python
tips.loc[tips['level']!='good',['tip','level']].head()
```

```python
# 새 컬럼 추가하여 데이터 만들기(반복문)
rows = tips.shape[0]
print(rows)
for i in range(rows):
    tips.loc[i,'index_id'] = str(1+i) #str은 문자열로 변환. 예 1 -> '1'
tips
```

```python
tips_m = tips.set_index('index_id')
tips_m
```

```python
# loc은 index 이름을 넣어줘야 함
tips_m.loc['2']
```

```python
# iloc은 index이름과 상관 없이 행의 고유 넘버를 넣어줘야 함
tips_m.iloc[2]
```

```python
tips.head()
```

```python
# loc은 마지막 값 포함
tips.loc[:1]
```

```python
# iloc은 마지막 값 미포함
tips.iloc[:1]
```

```python
# ratio 컬럼은 tip을 total_bill로 나눈 비율을 데이터로 함
tips['ratio'] = tips['tip']/ tips['total_bill']
tips.head()
```

```python
# 험수를 사용해서 새로운 컬럼 생성
# bill = totall_bill - tip
def bill(x) :
    total_bill = x['total_bill']
    tip = x['tip']
    bill = total_bill - tip
    return bill
tips['bill'] = tips.apply(bill, axis='columns')
tips
```

```python
# one-hot encoding과 빈 값 채우는 방법

tips['level_good'] = 0
tips['level_soso'] = 0
tips['level_bad'] = 0

tips.loc[tips['level'] == 'good', 'level_good'] =1
tips.loc[tips['level'] == 'soso', 'level_soso'] =1
tips.loc[tips['level'] == 'bad', 'level_bad'] =1

tips
```

```python
tips.describe()
```

```python
tips.loc[(tips['total_bill']>24)&(tips['tip']>3.5),'cusmomer'] = 'vip'
tips.loc[(tips['total_bill']<=24)&(tips['tip']<=3.5),'cusmomer'] = 'non-vip'
tips
```

```python
tips['total_bill(pre)'] = tips['total_bill'].shift(+1)
tips['total_bill(next)'] = tips['total_bill'].shift(-1)
tips[['total_bill','total_bill(pre)','total_bill(next)']]
```

```python
tips
```

```python
# 불필요한 컬럼 제거하기
#axis에서 1은 columns와 같과 0은 rows와 같음
tips_s = tips.drop(['index_id','total_bill(pre)','total_bill(next)'],axis='columns')
tips_s
```
